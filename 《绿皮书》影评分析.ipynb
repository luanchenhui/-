```python
# 构建网页信息获取函数
def get_data(urli):
    try:
        ri = requests.get(url = urli)
        ri.encoding = 'utf-8'
            # 访问网页
        soupi = BeautifulSoup(ri.text)
            # 解析网页
        infor_lst = soupi.find('div',id="comments").find_all('div',class_="comment-item")
            # 获取所有招聘标签div
        datalsti = []
        for infor in infor_lst[:]:
            dic = {}
            dic['评论者'] = infor.find('span',class_="comment-info").find('a').text
            dic['评分'] = int(infor.find('span',class_="comment-info").find_all('span')[1]['class'][0][-2:])
            dic['评论时间'] = infor.find('span',class_="comment-time").text.replace(' ','').replace('\n','')
            dic['点赞数量'] = int(infor.find('span',class_="votes").text)
            dic['评论内容'] = infor.find('p').text.replace('\n','')
            datalsti.append(dic)
        return datalsti
    except:
        return []

# u = 'https://movie.douban.com/subject/26100958/comments?start=0&limit=20&sort=new_score&status=P'
u1 = 'https://movie.douban.com/subject/27060077/comments?status=P'
get_data(u1)[:3]
```

```python
# 构建获取页面url函数
def get_urls(n):
    urllst = []
    for i in range(n):
        # urllst.append('https://movie.douban.com/subject/26100958/comments?start=%i&limit=20&sort=new_score&status=P' % (i*20))
        urllst.append('https://movie.douban.com/subject/27060077/comments?start=%i&limit=20&sort=new_score&status=P' % (i*20))
    return urllst

urllsts = get_urls(50)
urllsts[:2]
```

```python
# 批量获取数据

datalst = []
n = 1
for u in urllsts:
    datalst.extend(get_data(u))
    print('成功获取%i条数据' % (n*20))
    n += 1

datalst[:2]
```

```python
# 数据格式转换

df = pd.DataFrame(datalst)
df.iloc[:10]
```

```python
# 统计影评的词频
import jieba
import collections
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True)
# 如果在内容方面，根据其点赞数目进行加权，效果会如何。
text = ""
for idx,row in df.iterrows():
    text += '\n'+row['评论内容']
seg_list = jieba.cut(text, cut_all=False)
counter = collections.Counter(seg_list)
table = list(counter.items())
table.sort(key=lambda x:x[1], reverse=True)
x_words = list()
y_counts = list()
for row in table:
    if len(row[0]) > 1 and row[1] > 8:
        x_words.append(row[0])
        y_counts.append(row[1])

data = [go.Scattergl(x=x_words, y=y_counts)]
plotly.offline.iplot(data)
```

```python
# 绘制词云图
import matplotlib.pyplot as plt
from wordcloud import WordCloud

%matplotlib inline
wordcloud = WordCloud(
    font_path="/opt/conda/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/MicrosoftYaHei.ttf").generate(" ".join(x_words))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# 创建函数查找文字出现次数

def name_count(namei,s):
    n = 0
    for i in s:
        if namei in i:
            n += 1
    return n

name_count('黑人',df['评论内容'])
```

```python
namelst = ['黑人','南方','乡村','白人','平等','救赎','音乐']
            
for i in namelst:
    print(i,name_count(i,df['评论内容']))
```

```python
# 计算不同关键字出现频率

namelst = ['黑人','南方','乡村','白人','平等','救赎','音乐']

lst = []
for namei in namelst:
    lst.append({'关键词': namei,'出现频率': name_count(namei,df['评论内容'])})

lst
```
